```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense,Flatten,Input,Dropout,AveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Path to the directory containing the 'mask' and 'without mask' folders
data_dir = r'C:\Users\moham\Desktop\computer vision\mask_detction'

# Set up ImageDataGenerator for training and validation
train_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)  # Split the data into 80% train, 20% validation

# Load the training data
train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(224, 224),  # MobileNetV2 expects 224x224 images
    batch_size=32,
    class_mode='categorical',  # For multi-class classification
    subset='training'  # Set for training data
)

# Load the validation data
validation_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='validation'  # Set for validation data
)

# Check class labels
print(train_generator.class_indices)
```

```python
baseModel = MobileNetV2(weights = 'imagenet',include_top=False, input_tensor=Input(shape =(224,224,3)))
baseModel.summary()
```

```python
outlayer = baseModel.output
outlayer = AveragePooling2D(pool_size = (7,7))(outlayer)
outlayer = Flatten(name = "flatten")(outlayer)
outlayer = Dense(128,activation='relu')(outlayer)
outlayer = Dropout(0.5)(outlayer)
outlayer = Dense(2,activation="softmax")(outlayer)

model = Model(inputs = baseModel.input , outputs = outlayer)
```

```python
for layer in baseModel.layers :
    layer.trainable = False

# Compile the model
model.compile(
    optimizer='adam',  # You can also experiment with other optimizers like 'sgd'
    loss='categorical_crossentropy',  # For multi-class classification (2 classes in your case)
    metrics=['accuracy']  # Monitor the accuracy of the model during training
)

# Train the model
training = model.fit(
    train_generator,  # This comes from the earlier ImageDataGenerator
    validation_data=validation_generator,  # Also from ImageDataGenerator
    epochs=20,  # Set the number of epochs to train the model
    steps_per_epoch=len(train_generator)//32,  # Define how many batches per epoch
    validation_steps=len(validation_generator)//32  # Validation batches per epoch
)
```

```python
# Path to your test data directory containing 'mask' and 'without mask' folders
test_data_dir = r'C:\Users\moham\Desktop\computer vision\mask_detction'

# Initialize ImageDataGenerator for test data (no data augmentation)
test_datagen = ImageDataGenerator(rescale=1./255)

# Create a test data generator
test_generator = test_datagen.flow_from_directory(
    test_data_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=False  # Important for accurate predictions and evaluation
)
```

```python
# Evaluate the model on the test data
test_loss, test_accuracy = model.evaluate(test_generator, steps=len(test_generator))

print(f'Test Loss: {test_loss:.4f}')
print(f'Test Accuracy: {test_accuracy:.4f}')
```

```python
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix

# Predict the classes of the test data
Y_pred = model.predict(test_generator, steps=len(test_generator))
y_pred = np.argmax(Y_pred, axis=1)

# True labels
true_classes = test_generator.classes
class_labels = list(test_generator.class_indices.keys())

# Classification report
print('Classification Report')
print(classification_report(true_classes, y_pred, target_names=class_labels))

# Confusion matrix
print('Confusion Matrix')
print(confusion_matrix(true_classes, y_pred))
```
