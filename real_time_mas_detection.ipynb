```python
from tensorflow.keras.models import load_model
import numpy as np
import cv2
```

```python
model = load_model(r"C:\Users\moham\Desktop\computer vision\mask_detector_model.h5")
```

```python
cap = cv2.VideoCapture(0)

import numpy as np

def preprocess_frame(frame, target_size=(224, 224)):
    # Resize the frame to match the input size of the model
    resized_frame = cv2.resize(frame, target_size)
    # Normalize the pixel values to [0, 1]
    normalized_frame = resized_frame / 255.0
    # Reshape for CNN input
    reshaped_frame = np.expand_dims(normalized_frame, axis=0)
    return reshaped_frame
```

```python
while True:
    ret, frame = cap.read()  # Capture frame-by-frame
    if not ret:
        break

    # Preprocess the frame
    processed_frame = preprocess_frame(frame)

    # Predict mask or no mask
    prediction = model.predict(processed_frame)

    # Get class label based on the prediction
    mask_label = "Mask" if prediction[0][0] > 0.5 else "No Mask"

    # Display the label on the frame
    cv2.putText(frame, mask_label, (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Show the frame with label
    cv2.imshow('Mask Detection', frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and close windows
cap.release()
cv2.destroyAllWindows()
```
